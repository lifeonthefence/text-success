```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from datetime import datetime, timedelta
import time
```

```python
# Set search terms and date range
keyword = 'ukraine'
hashtag = ['#ukrainewar', '#UkraineRussiaWar️']#, '#putinwarcriminal', '#zelenskywarcriminal', '#standwithrussia', '#istandwithrussia', '#StandWithUkraine', '#IStandWithUkraine']
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)
lang_code = "en"
follower_threshold = 1000
tweet_limit = 100
```

```python
#the way I try to get the tweets per day
tweets = []
tweet_ids = set()
start_date = datetime(2022, 1, 1)
end_date = datetime(2022, 12, 31)
delta = timedelta(days=1)

while start_date <= end_date:
    # Set up date range for this day
    day_start = start_date.replace(hour=0, minute=0, second=0)
    day_end = start_date.replace(hour=23, minute=59, second=59)
    for tag in hashtag:
        query = f'{tag} lang:{lang_code} since:{day_start.date()} until:{day_end.date()}'
        counter = 0
        for tweet in sntwitter.TwitterSearchScraper(query).get_items():
            # fetch all results and filter by the user's followersCount
            if tweet.user.followersCount >= 1000:
                tweets.append({
                    'id': tweet.id,
                    'date': tweet.date,
                    'user': tweet.user.username,
                    'user_followers': tweet.user.followersCount,
                    'raw_content': tweet.content,
                    'rendered_content': tweet.renderedContent,
                    'likes': tweet.likeCount,
                    'retweets': tweet.retweetCount,
                    'replies': tweet.replyCount,
                    'quoteCount': tweet.quoteCount,
                    'hashtags': tweet.hashtags,
                    'lang': tweet.lang,
                    'media': tweet.media,
                    'mentionedUsers': tweet.mentionedUsers,
                })
                counter += 1
                if counter == 10:
                    break
    start_date += delta
```

```python
df = pd.DataFrame(tweets)
```
